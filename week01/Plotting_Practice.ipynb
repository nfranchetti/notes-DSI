```python
import matplotlib
import matplotlib.pyplot as plt    #import matplotlib libary
%matplotlib inline
```

```python
x = [1,2,3,4]                      #define some data
y = [20, 21, 20.5, 20.8]
plt.plot(x, y)                      #plot data
```

```python
# import matplotlib
# import matplotlib.pyplot as plt 
# %matplotlib inline # For use only in Jupyter Notebooks
fig = plt.figure() # Instantiates a figure, which will be empty
ax = fig.add_subplot(111) # Adds a subplot (read, axes) to the figure
ax.set(xlim=[0.5, 4.5], 
       ylim=[5, 32], 
       title='An Example Axes', 
       ylabel='Y-Axis', 
       xlabel='X-Axis') # Sets parameters for those axes (how far they go, their titles and labels)
ax.bar([0, 1, 2, 3], [15, 30, 6, 8], color='purple') # Plot 4 bars on Axes
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3) # Plot graph (default line) on Axes
ax.scatter([4.2, 3.8, 2.5, 3.5], [11, 25, 9, 26], color='darkgreen', marker='^') # Plot scatter points on Axes
plt.show()
```
