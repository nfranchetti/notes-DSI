# Case #1: Congressional Voting Data

After you've downloaded the data from the repository, go ahead and load it with Pandas

```python
df = pd.read_csv('../../DSI-CHI-1/lessons/week-07/2.3-pca-lab-1/assets/datasets/votes.csv', index_col=0)
print df.shape
df.head()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: 1 if x == 'y' else 0)
```

```python
df.head()
```

Next, let's define the x and y variables: 

```python
classes = pd.read_csv('../../DSI-CHI-1/lessons/week-07/2.3-pca-lab-1/assets/datasets/votes.csv', index_col=0)
y = classes['Class']
y.head()
```

```python
X = df[[col for col in df.columns if col != 'Class']]
X.head()
```

Next, create the covariance matrix from the standardized x-values and decompose these values to find the eigenvalues and eigenvectors

```python
X_standard = StandardScaler().fit_transform(X)
X_standard.shape
```

```python
covariance_matrix = np.cov(X_standard.T)
covariance_matrix.shape
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
```

Now, let's check the eigenvalues: 

```python
print eigenvalues.shape
eigenvalues
```

And the eigenvectors: 

```python
print eigenvectors.shape
eigenvectors
```

To find the principal componants, find the eigenpairs, and sort them from highest to lowest. 

```python
eigenpairs = zip(eigenvalues, eigenvectors)
eigenpairs
```

Now, calculate the explained variance. Recall the methods we learned in lesson 2.2!

```python
for i in sorted(eigenpairs, reverse=True):
    print i
```

Now, calculate the explained variance and the Cumulative explained variance

```python
eigensums = sum(eigenvalues)
eigensums
```

```python
for eig in sorted(eigenpairs, reverse=True):
    print eig[0] / eigensums, eig[0]
```

```python
np.cumsum([(eig / eigensums) * 100.0 for eig in sorted(eigenvalues, reverse=True)])
```

```python
# List comprehension version of the cell below

fit_me = [(eigenpair[1].reshape(16,1)) for eigenpair in sorted(eigenpairs, reverse=True)[0:8]]
```

```python
fit_me = []

for eigenpair in sorted(eigenpairs, reverse=True)[0:8]:
    fit_me.append(eigenpair[1].reshape(16,1))
```

```python
len(fit_me)
```

```python
Pmatrix = np.hstack(fit_me)
```

```python
Pmatrix.shape
```

```python
# Pmatrix = np.hstack((eigenpairs[0][1].reshape(4,1),
#                     eigenpairs[1][1].reshape(4,1)))
```

```python
pca_data = X_standard.dot(Pmatrix)
```

```python
pca_data
```

Now, conduct the PCA using scikit learn

```python
pca = PCA(n_components=8)
pca_x = pca.fit_transform(X_standard)
pca_x[0:5]
```

```python
from sklearn.cluster import KMeans

regular_model = KMeans(2)

pca_model = KMeans(2)
```

```python
def how_many_ones(x):
    ones = [y for y in x if y == 1]
    return len(ones) / float(len(x))
```

```python
results = regular_model.fit(X_standard)
how_many_ones(results.labels_)
```

```python
pca_results = pca_model.fit(pca_x)
how_many_ones(pca_results.labels_)
```
