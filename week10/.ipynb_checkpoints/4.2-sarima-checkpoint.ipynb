```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../../DSI-CHI-1/lessons/week-10/4.2-seasonal-arima-lab/assets/datasets/h02.csv', index_col=0, names=['sales'])
df.index = pd.to_datetime(df.index)
df.head()
```

```python
df['days'] = df.index.day
df['days'].value_counts()
```

```python
df.drop('days', axis=1, inplace=True)
df.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(df['sales'])
# plt.figure()
decomp.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df['sales'], lags=30)
```

```python
diff = df['sales'].diff(periods=1)[1:]

plot_acf(diff, lags=30)
```
