```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
%matplotlib inline

df = pd.read_csv('../../DSI-CHI-1/lessons/week-10/4.1-arima-tuning/assets/data/euretail.csv')
df = df.set_index(['Year'])
df = df.stack().reset_index()
df[0].plot(title='European Retail Trade')
```

```python
df['trade'] = df[0]
df['Quarter'] = df['level_1'].apply(lambda x: x.replace('tr', ''))

def manual_replace(string):
    replace_string = string.strip()
    if replace_string[1] == '1':
        return '-01'
    elif replace_string[1] == '2':
        return '-04'
    elif replace_string[1] == '3':
        return '-07'
    else:
        return '-10'
    
df['month'] = df['Quarter'].apply(lambda x: manual_replace(x))
df['date'] = df['Year'].astype(str) + df['month']
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
df = df[['date', 'trade']]

pd.to_datetime(df['date'])

df.set_index('date', inplace=True)
pd.to_datetime(df.index)
df.index
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime


decomposition = seasonal_decompose(df['trade'])  
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(15, 8)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

diff0 = df.diff(periods=4)[4:]
diff0.plot(title='European Retail Trade Differenced')
plot_acf(diff0, lags=30)
plot_pacf(diff0, lags=30)
```

```python
diff1 = diff0.diff()[1:]
diff1.plot(title='European Retail Trade Differenced Twice')
plot_acf(diff1, lags=30)
plot_pacf(diff1, lags=30)
```

```python
import statsmodels.api as sm
data = df.stack().values
model = sm.tsa.statespace.SARIMAX(data, order=(0,1,1), seasonal_order=(0,1,1,4))
results = model.fit()
results.summary()
```

```python
# Don't plot the first 5 values, to account for data loss when differencing (d=1 + D=5)
residuals = results.resid[5:]
plt.plot(residuals)

plot_acf(residuals, lags=30)
plot_pacf(residuals, lags=30)
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

lags = min(10, len(residuals)/5)

acorr_ljungbox(residuals, lags=lags)
```

```python
model = sm.tsa.statespace.SARIMAX(data, order=(0,1,3), seasonal_order=(0,1,1,4), trend='c')
res = model.fit()

preds = res.forecast(12)
fcast = np.concatenate((data, preds), axis=0)

plt.figure();
plt.plot(data, 'o' , fcast, 'r--');
```

```python
def test_sarima_models(order_tuple, seasonal_tuple):
    model = sm.tsa.statespace.SARIMAX(data, order=order_tuple, seasonal_order=seasonal_tuple)
    results = model.fit()
    print 'AIC is', str(results.aic)
    residuals = results.resid[5:]
    lags = min(10, len(residuals)/float(5))
    ljung_box_results = acorr_ljungbox(residuals, lags=lags)
    print 'Minimum p-value in Ljung box is', str(min(ljung_box_results[1]))
    plt.plot(residuals)
    plot_acf(residuals, lags=30)
    plot_pacf(residuals, lags=30)
```

```python
test_sarima_models((1,1,0),(1,1,0,4))
```

```python
test_sarima_models((1,1,1),(1,1,0,4))
```
